```python
pd.options.mode.chained_assignment = None 
data['Date'] = data['Date'].astype(str, copy = True)
data['Year'] = data['Date'].str[:4]
data['Month'] = data['Date'].str[4:6]
data['Day'] = data['Date'].str[6:]
# extract year, month, day from date
```

```python
data['ConfirmedCases'].fillna(0, inplace = True)
data['ConfirmedDeaths'].fillna(0, inplace = True)
# fill confirm case and confirm death with 0
```

```python
date13 = data[(data['Year'] == '2022') & (data['Month'] == '03') & (data['Day'] == '13')].index
data.drop(index = date13, inplace = True)
date14 = data[(data['Year'] == '2022') & (data['Month'] == '03') & (data['Day'] == '14')].index
date15 = data[(data['Year'] == '2022') & (data['Month'] == '03') & (data['Day'] == '15')].index
data.drop(index = date14, inplace = True)
data.drop(index = date15, inplace = True)
# drop March 13,14,15 2022 from dataframe
```
