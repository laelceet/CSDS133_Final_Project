```python
import pandas as pd
df = pd.read_csv('/Users/vihuynh1301/Downloads/final.csv')

#Ohio 2020 + 2021
USA_Graph = df[(df['CountryCode'] == 'USA') &(df['Year'] == 2020) & (df['Day'] == 28) & (df['RegionName'] == 'Ohio')]
#print(USA_Graph)

fig, ax1 = plt.subplots()
plt.gcf().set_size_inches(6, 4)
ax1.plot(USA_Graph.Month, USA_Graph.GovernmentResponseIndex, 'g', label='Gov')
ax2 = ax1.twinx()
ax2.plot(USA_Graph.Month, (USA_Graph.ConfirmedCases), 'b', label='Cases')

ax1.set_xlabel('Months')
ax1.set_ylabel('Government Response Index', color='g')
ax2.set_ylabel('Confirmed Cases', color='b')
plt.title('Ohio - USA Confirmed Cases vs GRI in 2020', color = 'black')

fig, ax1 = plt.subplots()
plt.gcf().set_size_inches(6, 4)
ax1.plot(USA_Graph.Month, USA_Graph.GovernmentResponseIndex, 'g', label='Gov')
ax2 = ax1.twinx()
ax2.plot(USA_Graph.Month, (USA_Graph.ConfirmedDeaths), 'r', label='Cases')

ax1.set_xlabel('Months')
ax1.set_ylabel('Government Response Index', color='g')
ax2.set_ylabel('Deaths', color='r')
plt.title('Ohio - USA Deaths Cases vs GRI in 2020', color = 'black')
plt.show()

USA_Graph1 = df[(df['CountryCode'] == 'USA') &(df['Year'] == 2021) & (df['Day'] == 28) & (df['RegionName'] == 'Ohio')]
#print(USA_Graph1)

fig, ax1 = plt.subplots()
plt.gcf().set_size_inches(6, 4)
ax1.plot(USA_Graph1.Month, USA_Graph1.GovernmentResponseIndex, 'g', label='Gov')
ax2 = ax1.twinx()
ax2.plot(USA_Graph1.Month, (USA_Graph1.ConfirmedCases), 'b', label='Cases')

ax1.set_xlabel('Months')
ax1.set_ylabel('Government Response Index', color='g')
ax2.set_ylabel('Confirmed Cases', color='b')
plt.title('Ohio - USA Confirmed Cases vs GRI in 2021', color = 'black')

fig, ax1 = plt.subplots()
plt.gcf().set_size_inches(6, 4)
ax1.plot(USA_Graph1.Month, USA_Graph1.GovernmentResponseIndex, 'g', label='Gov')
ax2 = ax1.twinx()
ax2.plot(USA_Graph1.Month, (USA_Graph1.ConfirmedDeaths), 'r', label='Cases')

ax1.set_xlabel('Months')
ax1.set_ylabel('Government Response Index', color='g')
ax2.set_ylabel('Deaths', color='r')
plt.title('Ohio - USA Deaths Cases vs GRI in 2021', color = 'black')

plt.show()
```
